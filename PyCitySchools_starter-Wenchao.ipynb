```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_school=school_data_complete['school_name'].value_counts().count()

# Calculate the total number of students
total_students=school_data_complete['Student ID'].count()

# Calculate the total budget
total_budget='${:,.2f}'.format(school_data_complete['budget'].unique().sum())

# Calculate the average math score
avg_math_score='{:,.2f}'.format(school_data_complete['math_score'].mean())

# Calculate the average reading score
avg_reading_score='{:,.2f}'.format(school_data_complete['reading_score'].mean())

# Calculate the average score
overall_average_score='{:,.2f}'.format(((school_data_complete['math_score']+school_data_complete['reading_score'])/2).mean())

# Calculate the percentage of students with a passing math score (70 or greater). Create a dataset with the passing math score.
math_passing_complete=school_data_complete.loc[school_data_complete['math_score'] >=70]
math_passing_rate='{:,.2f}'.format(len(math_passing_complete)/total_students*100)

# Calculate the percentage of students with a passing reading score (70 or greater). Create a dataset with the passing reading score.
reading_passing_complete=school_data_complete.loc[school_data_complete['reading_score'] >=70]
reading_passing_rate='{:,.2f}'.format(len(reading_passing_complete)/total_students*100)

# Calculate the overall passing rate
overall_passing_rate='{:,.2f}'.format((float(reading_passing_rate)+float(math_passing_rate))/2)

# Create dataframe district_table to hold the above results
district_table=pd.DataFrame({'Total Schools':total_school,'Total Students':total_students,'Total Budget':total_budget,
                             'Average Math Score':avg_math_score,'Average Reading Score':avg_reading_score,'% Passing Math':math_passing_rate,
                             '% Passing Reading':reading_passing_rate,'Overall Average Score':overall_average_score,'% Overall Passing Rate':overall_passing_rate},
                            index=[0])
district_table
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Groupby the school name for the three data sets
school_passing_analysis=school_data_complete.groupby('school_name')
school_reading_passing_analysis=reading_passing_complete.groupby('school_name')
school_math_passing_analysis=math_passing_complete.groupby('school_name')

# Obtain the School Type. The groupby results's objective would show a list type for each school type. Solved by the below code.
school_type=[]
for i in range(len(school_passing_analysis['type'])):
    school_type.append(school_passing_analysis['type'].unique()[i][0])

# Calculate the total students of each school
school_total_student=school_passing_analysis['Student ID'].count()

# Calculate the total budget of each school. Note that the budget at each row is the total budget of each school. 
school_total_budget=(school_passing_analysis['budget'].sum()/school_total_student)

# Calculate the student budget of each school.
per_student_budget=school_total_budget/school_total_student

# Calculate Average Math Score per school
school_math_score=school_passing_analysis['math_score'].mean()

# Calculate Average Reading Score per school
school_reading_score=school_passing_analysis['reading_score'].mean()

# Calculate % Passing Math, % Passing Reading, and Overall Passing Rate per school
school_math_passing=school_math_passing_analysis['Student ID'].count()/school_total_student*100
school_reading_passing=school_reading_passing_analysis['Student ID'].count()/school_total_student*100
school_overall_passing=(school_math_passing+school_reading_passing)/2

# Create dataframe school_table to hold the above results
school_table=pd.DataFrame({'School Type':school_type, 'Total Student':school_total_student,
                           'Total School Budget':school_total_budget.map("${:,.2f}".format), 'Per Student Budget':per_student_budget.map("${:.2f}".format),
                          'Average Math Score':school_math_score,'Average Reading Score':school_reading_score,
                          '% Passing Math':school_math_passing,'% Passing Reading':school_reading_passing,'Overall Passing Rate':school_overall_passing})
# Clean the dataframe
school_table.index.names = ['']
# Obtain the top performing school
school_table.sort_values('Overall Passing Rate',ascending=False).head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Groupby the dataset with grade and school name and put into a new dataset
grade_score_analysis=school_data_complete.groupby(['grade','school_name'])

# Assume how many grades in the dataset is unknown but known as ordinal numbers.
grades=school_data_complete['grade'].unique()
# Sorting the ordinal grades. Took  2 hours scratching head but really enjoy the logic behind python now.
def ordinal_digit(y):
    x=float(''.join(n for n in y if n.isdigit()))
    return x
sorted_grades=sorted(list(grades),key=ordinal_digit)

# Obtain the table for math scores per each school and grade 
grade_math_table=pd.DataFrame({})
for i in sorted_grades:
    grade_math_table[f'{i}']=grade_score_analysis['math_score'].mean()[i].map('{:.2f}'.format)
grade_math_table.index.names = ['']
grade_math_table
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# Obtain the table for reading scores per each school and grade 
grade_reading_table=pd.DataFrame({})
for i in sorted_grades:
    grade_reading_table[f'{i}']=grade_score_analysis['reading_score'].mean()[i].map('{:.2f}'.format)
grade_reading_table.index.names = ['']
grade_reading_table
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Input the data of the budget with numbers instead of the money string
school_table['Per Student Budget']=per_student_budget

# Create the spending column with bins
school_table['Spending Ranges (Per Student)']=pd.cut(school_table["Per Student Budget"], spending_bins, labels=group_names)

# Create the dataset by groupby with the spending
spending_analysis=school_table.groupby('Spending Ranges (Per Student)')

# Save the results and show the table
spending_analysis_table=pd.DataFrame({'Average Math Score':spending_analysis['Average Math Score'].mean(),
                                     'Average Reading Score':spending_analysis['Average Reading Score'].mean(),
                                     '% Passing Math':spending_analysis['% Passing Math'].mean(),
                                     '% Passing Reading':spending_analysis['% Passing Reading'].mean(),
                                     'Overall Passing Rate':spending_analysis['Overall Passing Rate'].mean()})
spending_analysis_table
```
