```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.columns
```

## Player Count

* Display the total number of players


```python
# Get the total players
total_players=pd.DataFrame({'Total Players':[len(purchase_data["SN"].unique())]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Obtain the number of unique items
n_uni_items=len(purchase_data["Item ID"].value_counts())

# Obtain the average price of unique items
avg_price='${:,.2f}'.format(purchase_data["Price"].mean())

# Obtain the number of purchases
n_purchases=len(purchase_data['SN'])

# Obtain the total revenue
total_r='${:,.2f}'.format(purchase_data["Price"].sum())

# Create the dataframe and display
purchasing_analysis=pd.DataFrame({'Number of Unique Items':[n_uni_items],'Average Price':[avg_price],
                                  'Number of Purchases':[n_purchases],'Total Revenue':[total_r]})
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Copy the dataset for this analysis
copy_data=purchase_data

# Create a dataset "player" with no duplicates players
player=copy_data.drop_duplicates(subset=['Gender','SN'])

# Groupby the player dataset with 'Gender'
gender_player=player.groupby(['Gender']).count()

# Calculate the number player of each gender
total_player=gender_player['SN'].sum()

# Calculate the percentage of player of each gender
gender_percentage=['{:,.2f}'.format(gender_player['SN'][0]/total_player*100),
                   '{:,.2f}'.format(gender_player['SN'][1]/total_player*100),
                   '{:,.2f}'.format(gender_player['SN'][2]/total_player*100),]

# Create the dataframe and display with order
gender_table=pd.DataFrame({'Total Count':gender_player['SN'],'Percentage of Players':gender_percentage})
gender_table.index.names = ['']
gender_table.sort_values('Total Count',ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create gender analysis dataset by groupby the purchase data with 'gender'
gender_analysis=purchase_data.groupby(['Gender'])

# Obtain the purchase count, average purchase price, and total purchase value from the data set
purchase_count=gender_analysis['SN'].count()
avg_p_price=gender_analysis['Price'].mean().map("${:.2f}".format)
total_p_val=gender_analysis['Price'].sum().map("${:.2f}".format)

# Calcuate the average total purchase per person for each gender
avg_p_person=(gender_analysis['Price'].sum()/gender_player['Item ID']).map("${:.2f}".format)

# Create the dataframe and display
purchasing_gender_table=pd.DataFrame({'Purchase Count':purchase_count,
                                      'Average Purchase Price':avg_p_price,
                                      'Total Purchase Value':total_p_val,
                                      'Avg Total Purchase per Person':avg_p_person})
purchasing_gender_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins and group labels for ages
bins=[0,9,14,19,24,29,34,39,1000]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

# Create the 'Age Range' column in player dataset
player['Age Range']=pd.cut(player["Age"], bins, labels=group_labels)

# Create the 'age_group' by groupby player dataset with Age Range column
age_group=player.groupby('Age Range')

# Obtain the Total Count of players for each age bin
age_t_count=age_group['SN'].count()

# Calculate the percentage of players for each age bin
age_player_percentage=(age_t_count/total_player*100).map("{:.2f}".format)

# Create the age demographics table dataframe and display
age_table=pd.DataFrame({'Total Count':age_t_count,'Percentage of Players':age_player_percentage})                                 
age_table.index.names = ['']
age_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create the 'Age Range' column in purchase data dataset
purchase_data['Age Range']=pd.cut(purchase_data["Age"], bins, labels=group_labels)

# Create the 'age_group' by groupby purchase data dataset with Age Range column
age_analysis=purchase_data.groupby('Age Range')

# Obtain purchase count, average purchase price, total purchase value
purchase_count=age_analysis['SN'].count()
age_total_purchase=age_analysis['Price'].sum().map("${:.2f}".format)
age_avg_pruchase=age_analysis['Price'].mean().map("${:.2f}".format)

# Calculate the average total purchase per person for each age bin
age_person_purchase=((age_analysis['Price'].sum())/age_t_count).map("${:.2f}".format)

# Create the age purchase analysis table dataframe and display
age_p_table=pd.DataFrame({'Purchase Count':purchase_count,'Average Purchase Price':age_avg_pruchase,
                         'Total Purchase Value':age_total_purchase,'Avg Total Purchase per Person':age_person_purchase})
age_p_table.index.names = ['']
age_p_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a 'spender_analysis' dataset by groupby purchase data dataset with 'SN' column for each player
spender_analysis=purchase_data.groupby('SN')

# Obtain purchase count and save into an new data set spender_purchase count
spender_purchase_count=spender_analysis.count()

# Obtain Total Purchase Value and put into a new column
spender_purchase_count['Total Purchase Value']=spender_analysis['Price'].sum()

# Sort the dataset with top 'Total Purchase Value'
spender_purchase_count=spender_purchase_count.sort_values('Total Purchase Value',ascending=False)

# Create the column for purchase count
spender_purchase_count['Purchase Count']=spender_purchase_count['Purchase ID']

# Calculate the average perchase price per player
spender_purchase_count['Average Purchase Price']=(spender_purchase_count['Total Purchase Value']/spender_purchase_count['Purchase Count']).map("${:.2f}".format)

# Create the top spender table dataframe and display with clean format
spender_purchase_count['Total Purchase Value']=spender_purchase_count['Total Purchase Value'].map("${:.2f}".format)
spender_analysis_table=spender_purchase_count[['Purchase Count','Average Purchase Price','Total Purchase Value']]
spender_analysis_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a 'item_analysis' dataset by groupby purchase data dataset with 'Item ID','Item Name' column
item_analysis=purchase_data.groupby(['Item ID','Item Name'])

# Create a 'item_purchase_count' dataset with sorted data with purchase count
item_purchase_count=item_analysis.count().sort_values('Purchase ID',ascending=False)

# Obtain the total purchase value and item price. Clean the format and put into corresponding columns with purchase count.
item_purchase_count['Total Purchase Value']=item_analysis['Price'].sum()
item_purchase_count['Purchase Count']=item_purchase_count['Purchase ID']
item_purchase_count['Item Price']=(item_purchase_count['Total Purchase Value']/item_purchase_count['Purchase Count']).map("${:.2f}".format)
item_purchase_count['Total Purchase Value']=item_purchase_count['Total Purchase Value'].map("${:.2f}".format)

# Create the top purchase count table dataframe and display
item_analysis_table=item_purchase_count[['Purchase Count','Item Price','Total Purchase Value']].head()
item_analysis_table
```
